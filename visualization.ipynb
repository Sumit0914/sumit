### customization of a plot

   1. change the color of the plot
   2. change the marker of the plot
   3. change the line styal
   4. Add a grid
   5. Add a legend 
   6. add a title and axis labels

```python
fig = plt.figure(figsize=(6,4),edgecolor='#9031dd',linewidth=1)
x = np.linspace(0,10,100) 
y = np.sin(x)
plt.plot(x,y)
```

```python
fig = plt.figure(figsize=(6,4))
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x,y, linestyle="--", color="blue", linewidth=1, label="sin(x)")
plt.title("time series")
plt.xlabel("time")
plt.ylable("value")
plt.legend()
```

```python
fig = plt.figure(figsize=(6,2))
x = np.linspace(0, 10, 100)
y = np.sin(x)
z = np.cos(x)
plt.plot(x,y, linestyle="--", color="blue", linewidth=1, label="sin(x)")
plt.plot(x,z, linestyle=":", color="red", linewidth=1, lable="cos(x)")
plt.title("time series", loc="left",pad=10, fontdict={"fontsize":16,"color":"blue","weight":"bold"})
plt.xlabel("time")
plt.ylabel("value")
plt.legend()
```

```python
fig = plt.figure(figsize=(6,4))
x = np.linspace(0, 10, 100)
y = np.sin(x)
z = np.cos(x)
plt.plot(x,y, linestyle="--", color="blue", linewidth=1, label="sin(x)")
plt.plot(x,z, linestyle=":", color="red", linewidth=1, lable="cos(x)")
plt.xlim(-1,12)
plt.ylim(-1.5,1.5)
```

```python
rainfall = [0.8, 0.6, 0, 0, 0.1, 0.2, 1.2, 1.5, 0.1, 0.3, 0.6, 0.8]
month = ["jan", "feb", "mar", "apr", "may", "jun", "jul", "Aug", "sep", "Oct", "Nov", "Dec"]
print(len(rainfall), len(month))
fig = plt.figure(figsize=(6, 4))
plt.bar(month, rainfall, color="blue", hatch="//")
plt.title("rainfall in 2019")
plt.xlabel("month")
plt.ylabel("rainfall (mm)")
```
